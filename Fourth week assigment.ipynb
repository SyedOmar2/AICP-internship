```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('./menu.csv')
```

```python
data.describe()
```

```python
data.columns
```

```python
x = ['Calories', 'Total Fat', 'Carbohydrates', 'Dietary Fiber', 'Sugars', 'Protein', 'Vitamin A (% Daily Value)','Vitamin C (% Daily Value)','Calcium (% Daily Value)', 'Iron (% Daily Value)']
data[x].max()
```

Q.3: Check to see if infact there is any correlation between Calories and other independent variables by
plotting a correlation matrix next.

```python
asa = data[x]
corr = asa.corr()
sns.heatmap(corr, annot=True)
plt.title('Correlation Matrix')
plt.show()
```

Draw boxplot for Calories vs Category to spot outliers and max calories category

```python
# Set the figure size
plt.figure(figsize=(7, 8))

# Draw the boxplot
sns.boxplot(x='Category', y='Calories', data=data)

# Rotate the x-axis labels for better readability
plt.xticks(rotation=90)

# Add labels and title
plt.xlabel('Category')
plt.ylabel('Calories')
plt.title('Boxplot of Calories by Category')

# Show the plot
plt.show()
```

5: Figure out which exact item contains a high quantity for ['Calories', 'Total Fat', 'Carbohydrates',
'Dietary Fiber', 'Sugars', 'Protein', 'Vitamin A (% Daily Value)', 'Vitamin C (% Daily Value)', 'Calcium
(% Daily Value)', 'Iron (% Daily Value)'. Display summary link this:

```python
# Calculate the maximum value for each attribute
max_values = data.max()

# Display the maximum values for each attribute
print("Maximum values for each attribute:")
print(max_values)
print()

# Identify the item(s) with maximum values for each attribute
high_calories_item = data[data['Calories'] == max_values['Calories']]
high_total_fat_item = data[data['Total Fat'] == max_values['Total Fat']]
high_carbohydrates_item = data[data['Carbohydrates'] == max_values['Carbohydrates']]
high_dietary_fiber_item = data[data['Dietary Fiber'] == max_values['Dietary Fiber']]
high_sugars_item = data[data['Sugars'] == max_values['Sugars']]
high_protein_item = data[data['Protein'] == max_values['Protein']]
high_vitamin_a_item = data[data['Vitamin A (% Daily Value)'] == max_values['Vitamin A (% Daily Value)']]
high_vitamin_c_item = data[data['Vitamin C (% Daily Value)'] == max_values['Vitamin C (% Daily Value)']]
high_calcium_item = data[data['Calcium (% Daily Value)'] == max_values['Calcium (% Daily Value)']]
high_iron_item = data[data['Iron (% Daily Value)'] == max_values['Iron (% Daily Value)']]

# Display summary linking to the exact item(s)
print("Item(s) with high quantities for each attribute:")
print("Calories:")
print(high_calories_item)
print()
print("Total Fat:")
print(high_total_fat_item)
print()
print("Carbohydrates:")
print(high_carbohydrates_item)
print()
print("Dietary Fiber:")
print(high_dietary_fiber_item)
print()
print("Sugars:")
print(high_sugars_item)
print()
print("Protein:")
print(high_protein_item)
print()
print("Vitamin A (% Daily Value):")
print(high_vitamin_a_item)
print()
print("Vitamin C (% Daily Value):")
print(high_vitamin_c_item)
print()
print("Calcium (% Daily Value):")
print(high_calcium_item)
print()
print("Iron (% Daily Value):")
print(high_iron_item)
```

```python
# Set the figure size
plt.figure(figsize=(12, 8))

# Draw horizontal bar graphs for each category
sns.barplot(x='Calories', y='Item', data=data[data['Category'] == 'Beef & Pork'], color='skyblue')

# Add labels and title
plt.xlabel('Calories')
plt.ylabel('Item')
plt.title('Calories in Items - Beef & Pork Category')

# Show the plot
plt.show()
```
