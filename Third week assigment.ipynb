```python
import matplotlib.pyplot as plt

# Set the x and y values for the line
x_values = [1, 2, 3, 4, 5]
y_values = [2, 4, 1, 4, 2]

# Create a line plot
plt.plot(x_values, y_values)

# Set the axis labels and title
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Sample graph!")

# Set the axis limits
plt.xlim(1, 3)
plt.ylim(1, 4)

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt


x1 = [1, 2, 3, 4, 5]
y1 = [2, 5, 1, 4, 2]


x2 = [1, 2, 3, 4, 5]
y2 = [3, 4, 2, 5, 3]


plt.plot(x1, y1, label='line1', marker='o')
plt.plot(x2, y2, label='line2', marker='s')


plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Line plot with markers")


plt.legend()


plt.show()
```

```python
import matplotlib.pyplot as plt

# Programming languages and their popularity
programming_languages = ["Java", "Python", "PHP", "JavaScript", "C#", "C++"]
popularity = [22.2, 17.6, 8.8, 8, 7.7, 6.7]

# Create the horizontal bar chart
plt.barh(programming_languages, popularity, color='skyblue')

# Add labels and title
plt.xlabel('Popularity')
plt.ylabel('Programming Languages')
plt.title('Popularity of Programming Languages')

# Show the chart
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample data
data = {
    'Fruits': ['apple', 'orange', 'banana', 'grapefruit', 'mango'],
    'Sales': [25, 30, 28, 32, 27]
}

# Load data into a DataFrame
df = pd.DataFrame(data)

# Create the bar plot
df.plot.bar(x='Fruits', y='Sales')

# Add labels and title
plt.xlabel('Fruits')
plt.ylabel('Sales')
plt.title('Fruit Sales')

# Rotate x-axis labels for better readability
plt.xticks(rotation=45)

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt

countries = ["United States", "Great Britain", "China", "Russia", "Germany"]
gold_medals = [46, 27, 26, 19, 17]


plt.pie(gold_medals, labels=countries, autopct="%1.1f%%")
plt.title("Gold Medal Achievements of Top 5 Countries in 2016 Summer Olympics")


plt.axis('equal')  


plt.show()
```

```python
import matplotlib.pyplot as plt

# Sample student data
math_marks = [88, 92, 80, 89, 100, 80, 60, 100, 80, 34]
science_marks = [35, 79, 79, 48, 100, 88, 32, 45, 20, 30]

# Create the scatter plot
plt.scatter(math_marks, science_marks)

# Add labels and title
plt.xlabel("Math Marks")
plt.ylabel("Science Marks")
plt.title("Comparison of Math and Science Marks")

# Add a grid for better readability
plt.grid(True)

# Show the plot
plt.show()
```
